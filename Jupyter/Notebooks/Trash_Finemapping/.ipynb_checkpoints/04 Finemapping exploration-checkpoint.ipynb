```python
import pandas as pd
import pyensembl
import os
import scipy.stats as stats
from plotnine import *
```

```python
dir='/nfs/research1/zerbino/jhidalgo/inteql_GTEX_v8/data/processed_data/gene_split_finemapping/'
```

```python
data=pd.DataFrame()
for i in os.listdir(dir+'eQTL/'):
    if i.startswith('ENSG'):
        gene=i.split('_')[0]
        eQTL=pd.read_csv(dir+'eQTL/'+gene+'_z.csv',sep='\t')
        if len(eQTL) > 1:
            eQTL['z']=stats.zscore(eQTL['pval_nominal'])
            eQTL.drop(columns=['pval_nominal']).to_csv(dir+'eQTL_z/'+gene+'_z.csv',index=False,sep='\t')
        #eQTL['rs_id_dbSNP151_GRCh38p7']=eQTL['rs_id_dbSNP151_GRCh38p7'].str.strip().astype(str)
        #pip=pd.read_csv(dir+'output/'+i,sep='\t',skiprows=1,names=['rs_id_dbSNP151_GRCh38p7','pip'])
        #pip['rs_id_dbSNP151_GRCh38p7']=pip['rs_id_dbSNP151_GRCh38p7'].str.strip().astype(str)
        #predata=eQTL.merge(pip)
        #predata['gene_id']=gene
        #data=pd.concat([data,predata])
```

```python
data=pd.DataFrame()
for i in os.listdir(dir+'output_2/'):
    if i.startswith('ENSG'):
        gene=i.split('_')[0]
        eQTL=pd.read_csv(dir+'eQTL_z/'+gene+'_z.csv',sep='\t')
        eQTL['rs_id_dbSNP151_GRCh38p7']=eQTL['rs_id_dbSNP151_GRCh38p7'].str.strip().astype(str)
        pip=pd.read_csv(dir+'output_2/'+i,sep='\t',skiprows=1,names=['rs_id_dbSNP151_GRCh38p7','pip'])
        pip['rs_id_dbSNP151_GRCh38p7']=pip['rs_id_dbSNP151_GRCh38p7'].str.strip().astype(str)
        predata=eQTL.merge(pip)
        predata['gene_id']=gene
        data=pd.concat([data,predata])
```
